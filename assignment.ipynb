```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data: property prices in thousands
neighborhood_A = [250, 300, 350, 400, 450, 500, 550, 600]
neighborhood_B = [260, 310, 360, 410, 460, 510, 560, 620]

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(neighborhood_A, neighborhood_B)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Visualize the data
plt.figure(figsize=(10, 6))
sns.histplot(neighborhood_A, kde=True, color='blue', label='Neighborhood A', bins=10)
sns.histplot(neighborhood_B, kde=True, color='green', label='Neighborhood B', bins=10)
plt.legend()
plt.title('Distribution of Property Prices in Neighborhood A and B')
plt.xlabel('Property Price (in thousands)')
plt.ylabel('Frequency')
plt.show()
```
